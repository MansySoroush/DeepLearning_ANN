```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pickle
```

```python
## Load the dataset
df = pd.read_csv("data/cleaned_data.csv")
df.head()
```

# Preprocessing Data

```python
## Encode categorical variables
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
df
```

```python
## One-hot Encoding Geography Column
one_hot_encoder_geo = OneHotEncoder()
geo_encoder = one_hot_encoder_geo.fit_transform(df[['Geography']]).toarray()
geo_encoder
```

```python
one_hot_geo_columns = one_hot_encoder_geo.get_feature_names_out(['Geography'])
one_hot_geo_columns
```

```python
geo_encoded_df = pd.DataFrame(geo_encoder, columns=one_hot_geo_columns)
geo_encoded_df
```

```python
## Combine one hot encoder columns with the original data
df = pd.concat([df.drop('Geography',axis=1), geo_encoded_df], axis=1)
df.head()
```

```python
## Save the encoders
with open('artifacts/label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('artifacts/one_hot_encoder_geo.pkl','wb') as file:
    pickle.dump(one_hot_encoder_geo,file)
```

```python
df.to_csv('artifacts/cleaned_preprocessed_data.csv', index=False)
```
